```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import mannwhitneyu, fisher_exact

def geo_mean(x):
    return np.exp(np.mean(np.log(x)))

with open('preprocessed/CUB_Conv4_proto_S_norm_ratio_1s5w.txt') as f:
    lines = f.readlines()
    
proto_S_norm_ratio = [float(line.split('\n')[0]) for line in lines]

with open('preprocessed/CUB_Conv4_proto_DR_norm_ratio_1s5w.txt') as f:
    lines = f.readlines()
    
proto_DR_norm_ratio = [float(line.split('\n')[0]) for line in lines]

proto_S_norm_ratio=np.array(proto_S_norm_ratio)
proto_DR_norm_ratio=np.array(proto_DR_norm_ratio)

print('Norm ratio')
print('Geometric means:',geo_mean(proto_S_norm_ratio),geo_mean(proto_DR_norm_ratio))
U1, p = mannwhitneyu(proto_S_norm_ratio, proto_DR_norm_ratio,alternative='two-sided') 
print('Mann–Whitney U test:',U1, p)
```

```python
with open('preprocessed/CUB_Conv4_proto_S_con_alpha_ratio_1s5w.txt') as f:
    lines = f.readlines()
    
proto_S_con_alpha_ratio = [float(line.split('\n')[0]) for line in lines]

with open('preprocessed/CUB_Conv4_proto_DR_con_alpha_ratio_1s5w.txt') as f:
    lines = f.readlines()
    
proto_DR_con_alpha_ratio = [float(line.split('\n')[0]) for line in lines]

proto_S_con_alpha_ratio=np.array(proto_S_con_alpha_ratio)
proto_DR_con_alpha_ratio=np.array(proto_DR_con_alpha_ratio)

print('Con-alpha ratio')
print('Geometric means:',geo_mean(proto_S_con_alpha_ratio),geo_mean(proto_DR_con_alpha_ratio))
print('Proportion',np.mean(proto_S_con_alpha_ratio<1),np.mean(proto_DR_con_alpha_ratio<1))
U1, p = mannwhitneyu(proto_S_con_alpha_ratio, proto_DR_con_alpha_ratio,alternative='two-sided')
print('Mann–Whitney U test:',U1, p)
table = np.array([[np.sum(proto_S_con_alpha_ratio<1), np.sum(proto_DR_con_alpha_ratio<1)], [np.sum(proto_S_con_alpha_ratio>=1), np.sum(proto_DR_con_alpha_ratio>=1)]])
oddsr, p = fisher_exact(table, alternative='two-sided')
print('Fisher\'s exact test:',oddsr, p)
```

```python
with open('preprocessed/CUB_Conv4_proto_S_div_alpha_ratio_1s5w.txt') as f:
    lines = f.readlines()
    
proto_S_div_alpha_ratio = [float(line.split('\n')[0]) for line in lines]

with open('preprocessed/CUB_Conv4_proto_DR_div_alpha_ratio_1s5w.txt') as f:
    lines = f.readlines()
    
proto_DR_div_alpha_ratio = [float(line.split('\n')[0]) for line in lines]

proto_S_div_alpha_ratio=np.array(proto_S_div_alpha_ratio)
proto_DR_div_alpha_ratio=np.array(proto_DR_div_alpha_ratio)

print('Div-alpha ratio')
print('Geometric means:',geo_mean(proto_S_div_alpha_ratio),geo_mean(proto_DR_div_alpha_ratio))
print('Proportion',np.mean(proto_S_div_alpha_ratio>1),np.mean(proto_DR_div_alpha_ratio>1))
U1, p = mannwhitneyu(proto_S_div_alpha_ratio, proto_DR_div_alpha_ratio,alternative='two-sided')
print('Mann–Whitney U test:',U1, p)
table = np.array([[np.sum(proto_S_div_alpha_ratio<1), np.sum(proto_DR_div_alpha_ratio<1)], [np.sum(proto_S_div_alpha_ratio>=1), np.sum(proto_DR_div_alpha_ratio>=1)]])
oddsr, p = fisher_exact(table, alternative='two-sided')
print('Fisher\'s exact test:',oddsr, p)
```
